```python
#Dependencies
import pandas as pd
from pathlib import Path

#Name the CSV files
schools_file = Path('C:\\Users\\stlki\OneDrive\\Desktop\\Homework\\pandas-challenge\\PyCitySchools\\Starter_Code\\PyCitySchools\\Resources\\schools_complete.csv')
students_file = Path('C:\\Users\\stlki\OneDrive\\Desktop\\Homework\\pandas-challenge\\PyCitySchools\\Starter_Code\\PyCitySchools\\Resources\\students_complete.csv')

#Create two data frames
schools_df = pd.read_csv(schools_file)
students_df = pd.read_csv(students_file)

#Combine the data sets
complete_school_data = pd.merge(students_df, schools_df, how = "left", on = ["school_name", "school_name"])
complete_school_data.head()
```

```python
#Calculate the total number of schools
school_count = len(complete_school_data["school_name"].unique())
school_count
```

```python
#Calculate the total number of students
student_count = len(complete_school_data["student_name"])
student_count
```

```python
#Calculate the total budget
total_budget = schools_df["budget"].sum()
total_budget
```

```python
#Calculate the average math score
average_math_score = complete_school_data["math_score"].mean()
average_math_score
```

```python
#Calculate the average reading score
average_reading_score = complete_school_data["reading_score"].mean()
average_reading_score
```

```python
#Calculate the percentage of students who passed math
passing_math_count = complete_school_data[(complete_school_data["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(student_count) * 100
passing_math_percentage
```

```python
#Calculate the percentage of students who passed reading
passing_reading_count = complete_school_data[(complete_school_data["reading_score"] >= 70)].count()["student_name"]
passing_reading_percentage = passing_reading_count / float(student_count) * 100
passing_reading_percentage
```

```python
#Calculate the percentage of students who passed both math and reading
passing_math_reading_count = complete_school_data[(complete_school_data["math_score"] >= 70) & (complete_school_data["reading_score"] >= 70)].count()["student_name"]
passing_math_reading_percentage = passing_math_reading_count / float(student_count) * 100
passing_math_reading_percentage
```

```python
#Summarize the data of the district
district_summary = pd.DataFrame({"Total Schools": [school_count],
                                 "Total Students": [student_count],
                                 "Total Budget": [total_budget],
                                 "Average Math Score": [average_math_score],
                                 "Average Reading Score": [average_reading_score],
                                 "% Passing Math": [passing_math_percentage],
                                 "% Passing Reading": [passing_reading_percentage],
                                 "% Overall Passing": [passing_math_reading_percentage]})
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("{:,.2f}".format)
district_summary
```

```python
#Declare the school types
school_types = schools_df.set_index(["school_name"])["type"]
```

```python
#Calculate the total student count per school
per_school_counts = schools_df.set_index(["school_name"])["size"]
```
